```python
title = placements.groupby('Industry').agg('count')
print(title)
```

As above the this table will show the sorted data from industry wise in our data

```python
ty_labels = title.Company.sort_values().index
ty_counts = title.Company.sort_values()
```

```python
plt.figure(1, figsize=(50,50)) 
the_grid = GridSpec(1,2 )
```

```python
cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
```

```python
plt.subplot(the_grid[0, 1], aspect=9, title='Industry Wise hiring in the college')
type_show_ids = plt.pie(ty_counts, labels=ty_labels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

As above we can see Software/IT Services hire the most students from our college


Education will be on the second in hiring the most students from college

```python
def glv(column):
    rg = placements.groupby(column).agg('count')
    pr = rg[lambda x: x.columns[0]].quantile(.75)
    v = rg[lambda x: x.columns[0]].loc[lambda s: s < pr].index.values
    def fix_values(row):
        if row[column] in v:
            row[column] = 'Other Idustries'
        return row 
    rg = placements.apply(fix_values, axis=1).groupby(column).agg('count')
    return rg
```

```python
rg = glv('Industry')
rg_labels = rg.Company.sort_values().index 
rg_counts = rg.Company.sort_values()
```

```python
plt.subplot(the_grid[0, 1], aspect=1, title='Industry Wise hiring in the college')
type_show_ids = plt.pie(rg_counts, labels=rg_labels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

As above it was difficult to see all industries so here we have calculate 75th percentile and industries having lower counts goes into 'Other Industries'

```python
def gl(column):
    rg = placements.groupby(column).agg('count')
    pr = rg[lambda x: x.columns[0]].quantile(.55)
    v = rg[lambda x: x.columns[0]].loc[lambda s: s < pr].index.values
    def fix_values(row):
        if row[column] in v:
            row[column] = 'Other Years'
        return row 
    rg = placements.apply(fix_values, axis=1).groupby(column).agg('count')
    return rg
```

```python
rg = gl('Year')
rg_labels = rg.Students_Selected.sort_values().index 
rg_counts = rg.Students_Selected.sort_values()
```

```python
plt.subplot(the_grid[0, 1], aspect=1, title='Selection of Students')
type_show_ids = plt.pie(rg_counts, labels=rg_labels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

-we have used 55 percentile in this chart

-As we can see in year 2021 and apart from 2022 with other years the hiring numbers were equvivalent
